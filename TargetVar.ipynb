```python
t2016 = cds_2016[['art_id', 'GrupoEditorial']]
t2017 = cds_2017[['art_id', 'GrupoEditorial']]
t2018 = cds_2018[['art_id', 'GrupoEditorial']]
```

```python
def target_var(target2017):
    if target2017.GrupoEditorial_x == target2017.GrupoEditorial_y :
        return (0)
    elif (target2017.GrupoEditorial_y == 90) and (target2017.GrupoEditorial_x != 90):
        return (1)
    else :
        return (0)

target2017['target_variable'] = target2017.apply(target_var, axis=1)
```

```python
Total = target2017['target_variable'].sum()
print (Total)
```
