```python
# Imports
import csv
import pandas as pd
import numpy as np
```

```python
# Reading CVS files for each year
cds_2016 = pd.read_csv('CDS_2016_va.csv',  encoding="ISO-8859-1")
cds_2017 = pd.read_csv('CDS_2017_va.csv',  encoding="ISO-8859-1")
cds_2018 = pd.read_csv('CDS_2018_va.csv',  encoding="ISO-8859-1")
cds_2019 = pd.read_csv('CDS_2019_NO_LABEL.csv',  encoding="ISO-8859-1")
```

```python
cds_2016.columns
```

```python
cds_2016.columns = ['Id_Cliente', 'Ano', 'Curso', 'Asignatura', 'TipoMaterial', 'GrupoEditorial', 'Lengua', 'TipoSoporte', 'Clase', 'ValorClase']
cds_2017.columns = ['Id_Cliente', 'Ano', 'Curso', 'Asignatura', 'TipoMaterial', 'GrupoEditorial', 'Lengua', 'TipoSoporte', 'Clase', 'ValorClase']
cds_2018.columns = ['Id_Cliente', 'Ano', 'Curso', 'Asignatura', 'TipoMaterial', 'GrupoEditorial', 'Lengua', 'TipoSoporte', 'Clase', 'ValorClase']
cds_2019.columns = ['Id_Cliente', 'Ano', 'Curso', 'Asignatura', 'TipoMaterial', 'Lengua', 'TipoSoporte', 'Clase', 'ValorClase']
```

```python
# Feature 'Curso' for the 2018 dataset has a 'c' character prior to each code. We will remove it and convert the column to integer type.
cds_2018['Curso'] = cds_2018['Curso'].replace('c', '', regex=True)
cds_2018['Curso'] = cds_2018['Curso'].astype(int)
```

```python
# Modifying the value for year in 2018 dataset:
cds_2018['Ano'] = 2018
```

We will create an artificial ID to identify unique combinations, based on the combination of the following features:
Id-Cliente, Curso, Asignatura, TipoMaterial, GrupoEditorial, Lengua, TipoSoporte

```python
cds_2016['art_id'] = cds_2016.apply(lambda row:str(row.Id_Cliente)+str(row.Curso)+str(row.Asignatura)+str(row.TipoMaterial)+str(row.Lengua)+str(row.TipoSoporte), axis=1)
cds_2016['art_id'] = cds_2016.apply(lambda row:str(row.Id_Cliente)+str(row.Curso)+str(row.Asignatura)+str(row.TipoMaterial)+str(row.Lengua)+str(row.TipoSoporte), axis=1)
cds_2016
```

```python
cds_2017['art_id'] = cds_2017.apply(lambda row:str(row.Id_Cliente)+str(row.Curso)+str(row.Asignatura)+str(row.TipoMaterial)+str(row.Lengua)+str(row.TipoSoporte), axis=1)
cds_2017['art_id'] = cds_2017.apply(lambda row:str(row.Id_Cliente)+str(row.Curso)+str(row.Asignatura)+str(row.TipoMaterial)+str(row.Lengua)+str(row.TipoSoporte), axis=1)
cds_2017
```

```python
cds_2018['art_id'] = cds_2018.apply(lambda row:str(row.Id_Cliente)+str(row.Curso)+str(row.Asignatura)+str(row.TipoMaterial)+str(row.Lengua)+str(row.TipoSoporte), axis=1)
cds_2018['art_id'] = cds_2018.apply(lambda row:str(row.Id_Cliente)+str(row.Curso)+str(row.Asignatura)+str(row.TipoMaterial)+str(row.Lengua)+str(row.TipoSoporte), axis=1)
cds_2018
```

```python
cds_2019['art_id'] = cds_2019.apply(lambda row:str(row.Id_Cliente)+str(row.Curso)+str(row.Asignatura)+str(row.TipoMaterial)+str(row.Lengua)+str(row.TipoSoporte), axis=1)
cds_2019['art_id'] = cds_2019.apply(lambda row:str(row.Id_Cliente)+str(row.Curso)+str(row.Asignatura)+str(row.TipoMaterial)+str(row.Lengua)+str(row.TipoSoporte), axis=1)
cds_2019
```

```python
# Creating DataFrames for each year, to analiyze the change of GrupoEditorial between years:
t2016 = cds_2016[['art_id', 'GrupoEditorial']]
t2017 = cds_2017[['art_id', 'GrupoEditorial']]
t2018 = cds_2018[['art_id', 'GrupoEditorial']]
```

```python
target2017 = (pd.merge(t2016, t2017, on='art_id'))
target2017
```

```python
target2018 = (pd.merge(t2017, t2018, on='art_id'))
target2018
```

```python
# Creating the Target variable for 2016-2017 and 2017-2018:

def target_var(dataframe):
    if (dataframe.GrupoEditorial_y == 90) and (dataframe.GrupoEditorial_x != 90):
        return (1)
    else :
        return (0)

target2017['target'] = target2017.apply(target_var, axis=1)
target2018['target'] = target2018.apply(target_var, axis=1)
```

```python
target2017 = target2017.drop(columns=['GrupoEditorial_x', 'GrupoEditorial_y'])
```

```python
target2018 = target2018.drop(columns=['GrupoEditorial_x', 'GrupoEditorial_y'])
```

```python
final_target2017 = (pd.merge(target2017, cds_2017, on='art_id'))
```

```python
final_target2018 = (pd.merge(target2018, cds_2018, on='art_id'))
```

```python
# Validating that the Target has been assigned correctly:
final_target2018.loc[final_target2018['art_id'] == '21066335821131']
```

Merging dataframes with target variable from different years:

```python
cds_final = (pd.merge(final_target2017, final_target2018, on=['art_id', 'Id_Cliente', 'Curso', 'Asignatura', 'TipoMaterial', 'Lengua', 'TipoSoporte'], how="outer"))
```

```python
cds_final
```

Mapping columns according to the year that each record belongs to, since we need an Outer join:

```python
cds_final['Ano_y'] = cds_final.apply(
    lambda row: row['Ano_x'] if np.isnan(row['Ano_y']) else row['Ano_y'],
    axis=1
)
```

```python
# Keeping Grupo Editorial from previous year, since it should have a high significance for the model:
cds_final['GrupoEditorial_x'] = cds_final.apply(
    lambda row: row['GrupoEditorial_y'] if np.isnan(row['GrupoEditorial_x']) else row['GrupoEditorial_x'],
    axis=1
)
```

```python
cds_final['Clase_y'] = cds_final.apply(
    lambda row: row['Clase_x'] if np.isnan(row['Clase_y']) else row['Clase_y'],
    axis=1
)
```

```python
cds_final['ValorClase_y'] = cds_final.apply(
    lambda row: row['ValorClase_x'] if np.isnan(row['ValorClase_y']) else row['ValorClase_y'],
    axis=1
)
```

```python
cds_final['target_y'] = cds_final.apply(
    lambda row: row['target_x'] if np.isnan(row['target_y']) else row['target_y'],
    axis=1
)
```

```python
cds_final = cds_final.drop(columns=['Ano_x', 'Clase_x', 'ValorClase_x', 'target_x'])
```

```python
cds_final.rename(columns={"GrupoEditorial_x": "GrupoEditorialPrevio", "GrupoEditorial_y": "GrupoEditorialActual", "target_y": "Target", "Ano_y": "Ano", "Clase_y": "Clase", "ValorClase_y": "ValorClase"}, inplace=True)
```

Adding client features to the final training set:

```python
maestros = pd.read_csv('Maestros.csv')
```

```python
maestros.sample(5)
```

```python
training = (pd.merge(cds_final, maestros, on='Id_Cliente', how="left"))
```

```python
training.isna().sum()
```

```python
training.sample(5)
```

```python
training['ComunidadAutonoma'].replace(['Cataluña','Comunidad Valenciana', 'Islas Canarias', 'La Rioja', 'Madrid', 'Navarra', 'País Vasco'],[1,2,3,4,5,6,7],inplace=True)
training['Titularidad'].replace(['Católico', 'Privado', 'Público'],[1,2,3],inplace=True)
```

```python
training.sample(5)
```

Changing the Target to 0, for rows that do not exist in 2018 and Grupo Editorial Previo is 90 and for rows that turned to 90 in 2017 and stayed as 90 in 2018:

```python
training['Target'] = training.apply(
    lambda row: 0 if (row['GrupoEditorialPrevio']==90 and row['GrupoEditorialActual']==90) else row['Target'],
    axis=1
)
```

```python
training['Target'] = training.apply(
    lambda row: 0 if (row['GrupoEditorialPrevio']==90 and row['Ano']==2017) else row['Target'],
    axis=1
)
```

#### Generating the final CSV file for Training the model:

```python
training.to_csv('final_training.csv', sep=',')
```
